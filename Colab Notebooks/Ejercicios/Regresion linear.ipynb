```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#funcion para calcular los coeficientes de regresion
def estimar_b0_b1(x, y):
  #obtenemos cantidad de registros
  n = np.size(x)

  #obtenemos la media
  m_x, m_y = np.mean(x), np.mean(y)

  #calcular sumatoria de XY y mi sumatoria XX
  sumatoria_xy = np.sum( (x-m_x) * (y-m_y) )
  sumatoria_xx = np.sum( x*(x-m_x) )

  #coeficientes de regresion
  b_1 = sumatoria_xy / sumatoria_xx
  b_0 = m_y - b_1 * m_x

  return (b_0, b_1)


```

```python
#funcion de graficado
def plot_regresion(x, y, b):
  plt.scatter(x, y, color='g', marker='o', s=30) #pintamos los puntos x,y en color verde

  y_pred = b[0] + b[1]*x
  plt.plot(x, y_pred, color="b") #pintamos la linea de x con la formula de la regresion lineal en azul
  
  #etiquetado
  plt.xlabel("x, Independiente")
  plt.ylabel("y, Dependiente")

  plt.show()
```

```python
def main():
  #dataset
  x = np.array([1,2,3,4,5])
  y = np.array([2,3,5,6,5])

  #obtenemos los coeficientes b0 y b1
  b = estimar_b0_b1(x,y)
  print("Los valores b0 = {}, b1 = {}".format(b[0], b[1]))

  #graficamos los vectores y los coeficientes
  plot_regresion(x,y,b)
```

```python
if __name__ == "__main__":
  main()
```
